# **Instrucciones de instalación de OpenCV**

Dando por hecho de que ya tienes instalado en tu máquina Python, hay tres formas básicas de instalar la librería OpenCV: 


1.   Con ***pip***
2.   Con ***conda***
3. Con ficheros ***.whl***

Si te encuentras en un ordenador con permisos de administrador, la forma más sencilla de hacer la instalación de OpenCV es:
1. En una consola o terminal de Windows escribimos:
 
 `pip install opencv-python`

Si por el contrario te encuentras en un ordenador en cliente donde *pip* está bloqueado pero el paquete Anaconda tiene un proxy configurado, puedes instalar de la siguiente forma: 
2. En una consola o terminal de Windows escribimos:

 `conda install -c conda-forge opencv`

 2.1. [OPCIONAL] Si queremos crear un nuevo entorno de Conda e intalar OpenCV allí, podemos seguir los siguientes pasos: 

  2.1.1. Actualiza conda a la última versión disponible y crea un entorno desde la ventana de comandos o terminal:

   `conda update conda`
  
   `conda create --name opencv_env python==3.5.0`
  
  2.1.2. Activación de entorno:

   `activate opencv_env` 
  
  2.1.3. Instalación de OpenCV junto a sus dependencias (mínimas)

   `conda install -c menpo opencv3=3.1`

   `conda install jupyter matplotlib`
   

3. Por último, si ninguna de las dos formas anteriores son posibles, tendrás que buscar en internet los archivos *.whl* de OpenCV. En cliente suele funcionar el siguiente enlace donde encontraréis muchos de estas librerías para instalarlas: [https://www.lfd.uci.edu/~gohlke/pythonlibs/](https://www.lfd.uci.edu/~gohlke/pythonlibs/). Una vez descargado el fichero adecuado (en función de la versión de Python que tengáis instalada y si es de 32 o 64 bits) tendréis que escribir en la consola de comandos de Windows 

  `pip install ruta_del_whl_que_habéis_descargado`, por ejemplo en mi caso, donde mi versión de Python sería la 3.7 y la plataforma sería de 64 bits: : 
  
  `pip install "C:\Users\jpcb9\Downloads\opencv_python-4.4.0-cp37-cp37m-win_amd64.whl"`

 

Una forma de comprobar que la instalación de OpenCV ha funcionado correctamente sería abrir Python en nuestro terminal de Windows y escribir: 

 `python` 

Nos saldrá algo parecido a esto si tenemos Python en las variables de entorno: 

 `Python 3.7.3 (default, Mar 27 2019, 17:13:21) [MSC v.1915 64 bit (AMD64)] :: Anaconda, Inc. on win32`

 Y una vez aquí escribiremos: 

  `import cv2`

  Si no nos sale ningún mensaje de error, ¡enhorabuena! Has conseguido instalar OpenCV correctamente y estamos listos para hacer cosas divertidas con Python.  


# **Cargar imágenes y modificarlas**



> **NOTA IMPORTANTE:** Ejecutar todo este código en vuestro IDE favorito (Spyder, PyCharm o cualquier otro). Ya que muchas de las funcionalidades de OpenCV no están disponibles en Jupyter Notebook. *italicised text*



Antes de empezar nuestro código tenemos que importar la librería OpenCV, para ello usamos las expresión: 

`import cv2`

Luego, para cargar una imagen tendremos que utilizar la función ***cv2.imread()***, dándole como argumento la ruta de la imagen que queremos cargar. 
Opcionalmente le podemos dar un segundo argumento a la función que especifica la forma en la que la imagen debe ser leída.



*   cv2.IMREAD_COLOR: Opción por defecto. 
*   cv2.IMREAD_GRAYSCALE: Carga la imagen en modo escala de grises.
* cv2.IMREAD_UNCHANGED: Carga la imagen incluyendo el canal alpha (transparencias)


**Ejemplo 1.** Cargar una imagen y guardarla con otro nombre

```
### NOTA IMPORTANTE ### Copiar y pegar el código en Spyder o PyCharm y sustituir las rutas de las imágenes por las vuestras. 
import cv2

# Cargamos una imagen
image_path = r"C:\Users\jpcb9\Pictures\OpenCV_intro\myID.jpg"
image = cv2.imread(image_path)

# La guardamos pero ahora como png
image_copy_path = r"C:\Users\jpcb9\Pictures\OpenCV_intro\myID_copy.png"
cv2.imwrite(image_copy_path, image)

# Cargamos ahora la segunda imagen 
image_copy = cv2.imread(image_copy_path)

# Volvemos ahora a cargar la segunda imagen pero en blanco y negro
image_copy_byn = cv2.imread(image_copy_path, cv2.IMREAD_GRAYSCALE)

# Las mostramos en pantalla con la función cv2.imshow
cv2.imshow('ID original de Bender', image)
cv2.imshow('ID Copia', image_copy)
cv2.imshow('ID Copia en blanco y negro', image_copy, cv2.IMREAD_GRAYSCALE)


cv2.waitKey(0) # Esto esperará a que pulsemos una tecla para cerrar las imágenes
cv2.destroyAllWindows()
```

La imagen que yo he utilizado es ésta: 

myID.jpg
